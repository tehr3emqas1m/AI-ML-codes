# Linear Regression Example

```python
dict1 = {'F1':[20,25,28,29],'F2':[40,41,46,53]}
import pandas as pd
df = pd.DataFrame(dict1)
df
```

```python
from matplotlib import pyplot as plt
plt.scatter(df.F1,df.F2)
```

```python
df.corr()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = df[['F1']]
#X = np.array(df['F1'])
#X=X.reshape(-1,1)
y = df['F2']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

```python
X
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Line = [(x*model.coef_[0]) + model.intercept_ for x in df.F1]
```

```python
Line
```

```python
plt.scatter(df.F1,df.F2)
plt.plot(df.F1,Line,'r')
```
