#Data  pre-processing
Data Scaling
* Min MAx scaler

```python
Dict1 = {'Age':[30,25,27,33,41,23,27,29,33,34],
         'Saalary':[50000, 60000, 75000,55000,
                    65000,70000,80000,30000,90000,
                    71000]}
```

```python
import pandas as pd
df = pd.DataFrame(Dict1)
df.head()
```

```python
L_age=[]
L_sal=[]
for i in range(len(df)):
  L_age.append( (df['Age'][i]-df.Age.min())/ (df.Age.max()-df.Age.min())      )
  L_sal.append( (df['Saalary'][i]-df.Saalary.min())/ (df.Saalary.max()-df.Saalary.min())      )

```

```python
df_more_cols = df.copy()
df_more_cols['Age_normalized']=L_age
df_more_cols['Salary_normalzied'] = L_sal
df_more_cols
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
new_data = mms.fit_transform(df)
```

```python
new_data
```

```python
new_data_df = pd.DataFrame(new_data)
new_data_df.columns = ['Age_scaled','Saalry_scaled']
```

```python
new_data_df
```

* Now implementing z score for standardization

```python
df_new2 = df.copy()
```

```python

L_age_z=[]
L_sal_z = []
for i in range(len(df)):
  L_age_z.append((df_new2.Age[i]- df_new2.Age.mean())/df_new2.Age.std() )

  L_sal_z.append((df_new2.Saalary[i]- df_new2.Saalary.mean())/df_new2.Saalary.std() )

```

```python
df_new2['Age_new']=L_age_z
df_new2['Sal_new']=L_sal_z
df_new2
```
